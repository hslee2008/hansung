```python

from collections import Counter

# Function to extract "인문" and "자연" occurrences from texts


def extract_category_counts(text):
    humanities_count = text.count("인문")
    science_count = text.count("자연")
    return humanities_count, science_count


# Extract counts from before and after texts
before_humanities, before_science = extract_category_counts(before_text)
after_humanities, after_science = extract_category_counts(after_text)

# Calculate acceptance rates
humanities_acceptance_rate = (
    after_humanities / before_humanities) * 100 if before_humanities > 0 else 0
science_acceptance_rate = (
    after_science / before_science) * 100 if before_science > 0 else 0

# Create DataFrame for results
acceptance_df = pd.DataFrame({
    "Category": ["인문", "자연"],
    "Before": [before_humanities, before_science],
    "After": [after_humanities, after_science],
    "Acceptance Rate (%)": [humanities_acceptance_rate, science_acceptance_rate]
})


# Display the results
print(acceptance_df)

# Bar plot of Before and After Counts by Category
plt.figure(figsize=(10, 6))
acceptance_df.plot(x="Category", y=["Before", "After"], kind="bar", color=["skyblue", "orange"])
plt.title("Before and After Counts by Category")
plt.xlabel("Category")
plt.ylabel("Count")
plt.grid(axis="y", linestyle="--", alpha=0.7)
plt.show()
```
