```python
import requests
import xml.etree.ElementTree as ET
import pandas as pd
import matplotlib.pyplot as plt
from tqdm import tqdm
from PyPDF2 import PdfReader
import koreanize_matplotlib

# Function to extract school names from a PDF file
def extract_school_names_from_pdf(pdf_path):
    reader = PdfReader(pdf_path)
    text = "\n".join([page.extract_text() for page in reader.pages if page.extract_text()])
    lines = text.split("\n")

    school_names = []
    for line in lines:
        parts = line.split()
        if len(parts) > 1:  # Assuming school name is in the second column
            school_names.append(parts[1])

    return school_names

# Function to fetch school type from NEIS API with error handling
def get_school_type(school_name):
    url = f"https://open.neis.go.kr/hub/schoolInfo?SCHUL_NM={school_name}"
    response = requests.get(url)

    if response.status_code == 200:
        root = ET.fromstring(response.content)
        hs_sc_nm = root.find(".//HS_SC_NM")
        if hs_sc_nm is not None and hs_sc_nm.text:  # Check if element exists and is not None
            return hs_sc_nm.text.strip()
    return "Unknown"  # Default value if API fails or data is missing

# Extract school names from PDFs
interview_schools = extract_school_names_from_pdf("제12기 한성 노벨 영ㆍ수재 장학생 면접대상자 명단.pdf")
accepted_schools = extract_school_names_from_pdf("제12기 한성 노벨 영·수재 장학생 합격자 명단.pdf")

# Count occurrences of each school
interview_counts = pd.Series(interview_schools).value_counts().to_dict()
accepted_counts = pd.Series(accepted_schools).value_counts().to_dict()

# Create DataFrame
df_counts = pd.DataFrame({'School': list(set(interview_counts.keys()).union(set(accepted_counts.keys())))})

# Map interview and accepted counts
df_counts['Interview'] = df_counts['School'].map(interview_counts).fillna(0).astype(int)
df_counts['Accepted'] = df_counts['School'].map(accepted_counts).fillna(0).astype(int)

# Fetch school types with progress bar
df_counts["School Type"] = [
    get_school_type(school) for school in tqdm(df_counts["School"], desc="Fetching School Types")
]

# Aggregate by school type
df_grouped = df_counts.groupby("School Type")[["Interview", "Accepted"]].sum()

# Plot the data
fig, ax = plt.subplots(figsize=(12, 6))
df_grouped.plot(kind="bar", ax=ax)
ax.set_ylabel("참여자 수")
ax.set_xlabel("학교 유형")
ax.set_title("학교 유형별 선발 전후 참가자 비교")
plt.xticks(rotation=45)
plt.legend(["면접 대상자", "최종 합격자"])
plt.tight_layout()

plt.show()
```
